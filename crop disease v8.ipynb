```python
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os
import time
from PIL import Image
```

```python
model = YOLO(r"/Users/thamaraiselvi/Downloads/best3.pt")
```

```python
image_path = '/Users/thamaraiselvi/Downloads/test0.png'
image = cv2.imread(image_path)
results = model(image)
for r in results:
    boxes = r.boxes
    labels = r.names
    name = str(list(labels.values()))
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        w, h = x2-x1, y2-y1
        cv2.rectangle(image, (x1,y1),(x2,y2), (0,0,255), 10)


face = image[y1:y2, x1:x2]
plt.imshow(face[:,:,::-1])
face.shape
```
